```python
#importing libraries
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
#Importing dataset
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#Checking if there are any missing values present.
df.isnull().sum()
```

```python
#Spliting the data into predictors and target features
x=df.iloc[:,3:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
#encoding categorical features
#Label encoding gender feature
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['Gender']=le.fit_transform(x['Gender'])
```

```python
x.head()
```

```python
x['Geography'].unique()
```

```python
#One Hot Encoding the Geography feature
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x=pd.DataFrame(ct.fit_transform(x))
```

```python
x.head()
```

```python
#Spliting the dataset into training and testing data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,test_size=0.25,random_state=0)
```

```python
x_train.head()
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=pd.DataFrame(sc.fit_transform(x_train))
x_test=pd.DataFrame(sc.transform(x_test))
```

```python
x_train
```

#### Building the ann

```python
#Initializing the Ann
ann=tf.keras.models.Sequential()
```

```python
#Adding the input layer and the first hidden layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
#Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
#Adding the output layer
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

### Training the Ann

```python
#Compile the Ann
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Training the ANN on the trainig dataset
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
#Making Prediction and evaluating the model.
print(ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]]))>0.5)
```

```python
y_pred=ann.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
#Creating the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Accuracy Score
from sklearn.metrics import accuracy_score
a_s=accuracy_score(y_test,y_pred)
a_s
```
